## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
   
        

def draw_lines(img, lines, extrapolate, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness=3)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap, extrapolate):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if extrapolate == True:
        extrapolate_lines(line_img,lines)
    else:        
        draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., γ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + γ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, γ)
````

````python
def classify_lanes(lines):
    """
    This function classifies hough lines into two groups using slope.
    """
    left_x_coordinates = []
    left_y_coordinates = []
    right_x_coordinates = []
    right_y_coordinates = []    
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = ((y1 - y2) / (x2 - x1)) #write reason
            if slope > 0.3:
                left_x_coordinates.extend((x1,x2))
                left_y_coordinates.extend((y1,y2))
            elif slope < -0.3:
                right_x_coordinates.extend((x1,x2))
                right_y_coordinates.extend((y1,y2))                
    return (left_x_coordinates,left_y_coordinates), (right_x_coordinates, right_y_coordinates)
````

````python
def sketch_lanes(img, lane_coordinates, lane_line_buffer, color=[255, 0, 0], thickness=2):
    """
    This function sketchs the extrapolated lines on the image.
    """
    try:
        lane_equation = np.polyfit(lane_coordinates[0], lane_coordinates[1], 1) # find the line equation of the lane
        lane_line_buffer.append(lane_equation)    
    except TypeError:    
        pass
    
    # find the average of the last n lane line equations
    lane_line = [sum(x) / len(lane_line_buffer) for x in zip(*lane_line_buffer)] 
    
    
    y_intercept_start = img.shape[0]
    y_intercept_end = 0.625*img.shape[0]
    
    # find the intersection of the lane equation and a given line
    x_intercept_start = np.roots(np.array(lane_line)- [0,y_intercept_start]) 
    x_intercept_end = np.roots(np.array(lane_line)- [0,y_intercept_end])
    
    # add the new coordinates to the existing array
    lane_coordinates[0].extend((x_intercept_start, x_intercept_end))
    
    draw_x = lane_coordinates[0]
    draw_y = np.polyval(lane_line, draw_x)   # evaluate the polynomial

    draw_points = (np.asarray([draw_x, draw_y]).T).astype(np.int32)
    
    cv2.polylines(img, [draw_points], False, color, thickness) # sketch the extrapolated lane onto the image 
````

````python
def extrapolate_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    This function extrapolates the hough lines as complete lanes.
    """
    left_lane_coordinates, right_lane_coordinates = classify_lanes(lines)
    sketch_lanes(img, left_lane_coordinates, left_lane_line_buffer, color, thickness)
    sketch_lanes(img, right_lane_coordinates, right_lane_line_buffer, color, thickness)
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.

# Lane Finding Pipeline for a single image

# Step 1: Select the color of interest.
# Step 2: Select the region of interest.
# Step 3: Apply gaussian filter to smoothen the images.
# Step 4: Find edges using canny edge detector.
# Step 5: Create lines from edge pixels using Hough transform.
# Step 6: Superimpose the extrapolated lines on original image


def lane_finding_pipeline(original_image, debug=False):
    
    # Get the shape of the image
    shape = original_image.shape
   

    # Step 1: Select the color of interest.    
    COLOR_MIN = np.array([19, 0, 180],np.uint8) # minimum threshold for HSV image space
    COLOR_MAX = np.array([179, 255, 255],np.uint8)   
    
    original_image_hsv = cv2.cvtColor(original_image,cv2.COLOR_RGB2HSV) # convert the image from RGB to HSV color space
    color_selected_image = cv2.inRange(original_image_hsv, COLOR_MIN, COLOR_MAX) # select the color of interest
    

    # Step 2: Select the region of interest.
    vertices = np.array([[(0, shape[0]), ( math.ceil(shape[1]*0.48) , math.ceil(0.625*shape[0])), \
                          ( math.ceil(0.55*shape[1]) , math.ceil(0.625*shape[0])), (shape[1],shape[0])]], dtype=np.int32)
    masked_image = region_of_interest(color_selected_image, vertices)
    
    
    # Step 3: Apply gaussian filter to smoothen the images.
    gaussian_image = gaussian_blur(masked_image, kernel_size=5)

    # Step 4: Find edges using canny edge detector.
    edge_image = canny(gaussian_image, low_threshold=30, high_threshold=100)
    
    # Step 5: Create lines from edge pixels using Hough transform and extrapolate the lines
    lane_lines = hough_lines(edge_image, rho=1, theta=np.pi/180, threshold=30, \
                             min_line_len=5, max_line_gap=10, extrapolate=True)


    # Step 6: Superimpose the extrapolated lines on original image
    if debug == True:
        output_region_marked = cv2.polylines(lane_lines, vertices, True, [0,0,255], thickness=5)
        output_image = weighted_img(output_region_marked,original_image,α=.8, β=1.)
    else:
        output_image = weighted_img(lane_lines,original_image,α=.8, β=1.)
    
    if debug == True:
        # Print and save the results for debugging
        
        plt.imshow(color_selected_image,cmap='gray')
        plt.title('Step 1: Results after color selection')
        plt.imsave('test_pipeline_results/color_selected.jpg', color_selected_image, cmap='gray')
        plt.show()
        
        
        
        plt.imshow(masked_image, cmap='gray')
        plt.title('Step 2: Results after selecting the region of interest')
        plt.imsave('test_pipeline_results/region_selected.jpg', masked_image, cmap='gray')
        plt.show()  
        
        
        plt.imshow(gaussian_image, cmap='gray')
        plt.title('Step 3: Results after applying Gaussian Blur')
        plt.imsave('test_pipeline_results/gauss_filtered.jpg', gaussian_image, cmap='gray')
        plt.show()        
        
        
        plt.imshow(edge_image, cmap='gray')
        plt.title('Step 4: Results from Canny Edge Detector')
        plt.imsave('test_pipeline_results/canny_detected.jpg', edge_image, cmap='gray')
        plt.show() 
        
        
        plt.imshow(lane_lines, cmap='gray')
        plt.title('Step 5: Results from Hough Transform')
        plt.imsave('test_pipeline_results/hough_results.jpg', lane_lines, cmap='gray')
        plt.show()
        
        
        plt.imshow(output_image)
        plt.title('Step 6: Result after extrapolating the lane lines')
        plt.imsave('test_pipeline_results/output_extrapolated.jpg', output_image)
        plt.show() 
        
    return output_image
````
